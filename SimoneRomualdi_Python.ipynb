```python
# Struttura dati per conservare gli elementi
elementi = []

# Funzione per inserimento elementi
def inserimento_elementi():

    while True: # ciclo infinito
        nome = input('Inserisci nome: ')
        categoria = input('Inserisci categoria: ')

        # Controllo dell'input corretto del prezzo e quantità
        while True:
            try:
                prezzo = float(input('Inserisci prezzo: '))
                break # Si interrompe il ciclo While
            except ValueError: #Errore generato nella mancata conversione di un dato in un tipo non adatto
                print('Errore" Inserisci un numero valido: ')

        while True:
            try:
                quantità = int(input('Inserisci quantità: '))
                break
            except ValueError:
                print('Errore" Inserisci un numero valido per la quantità: ')

        condizione = input('Qual è la condizione? (scaduto, difetti estetici ecc..): ')

        scadenza = None # Valore di partenza
        anno_corrente = 2025 # Variabile per rendere più dinamico il controllo successivo sulle date di scadenza

        if 'scaduto' in condizione.lower():  # Per rendere il controllo case-insensitive
            while True:
                scadenza = input('Quando scade il prodotto? (Formato GG-MM-AAAA): ')

                # Controllo formato GG-MM-AAAA
                if len(scadenza) == 10 and scadenza[2] == '-' and scadenza[5] == '-' and scadenza.replace('-', '').isdigit(): # scadenza.replace toglie i '-' così da controllare che il resto siano numeri
                    giorno, mese, anno = map(int, scadenza.split('-'))  # Converte i valori in interi trasformando la data in lista così da evitare gli spazi

                    if 1 <= giorno <= 31 and 1 <= mese <= 12 and anno <= anno_corrente:
                        break
                    else:
                        print('Errore! La data deve essere nel passato o oggi.')

                else:  # Caso in cui il prodotto non sia scaduto
                    while True:
                        scadenza = input('Il prodotto non è scaduto, quando scade? (Formato GG-MM-AAAA, lascia vuoto se non applicabile): ').strip()

                if scadenza == '':  # Se l'utente non mette nulla
                    scadenza = None
                    break

                if len(scadenza) == 10 and scadenza[2] == '-' and scadenza[5] == '-' and scadenza.replace('-', '').isdigit():
                    giorno, mese, anno = map(int, scadenza.split('-'))

                    if 1 <= giorno <= 31 and 1 <= mese <= 12 and anno >= anno_corrente:
                        break
                    else:
                        print('Errore! Inserisci una data futura valida.')
                else:
                    print('Errore! Inserisci la data nel formato corretto.')

        # Creazione dizionario elementi
        cibo = {
            'nome' : nome,
            'categoria' : categoria,
            'prezzo' : prezzo,
            'quantità' : quantità,
            'scadenza' : scadenza,
            'condizione' : condizione
        }

        # Aggiungere alla lista il dizionario
        elementi.append(cibo)
        print(f'{nome} aggiunto correttamente')

        # Richiesta di aggiunta per un altro elemento.
        # Qualsiasi risposta diversa da 'Si' blocca il loop

        elemento_aggiuntivo = input('Vuoi aggiungere un altro elemento? (si/no): ').strip().lower()
        if elemento_aggiuntivo not in ['si', 'sì']: # per evitare problemi se il si viene scritto in maniera diversa
            break

# Funzione per visualizzazione elementi
def visualizza_elementi():

    # così che se la lunghezza è 0 vuol dire che non ci sono cibi inseriti nella lista
    if len(elementi) == 0:
        print('Non ci sono elementi registrati')
    else:
        for cibo in elementi: #uso di \n per andare a capo e di f-string per rendere il codice dinamico e più leggibile
            print('\nDettagli cibo')
            print(f"Nome: {cibo['nome']}")
            print(f"Categoria: {cibo['categoria']}")
            print(f"Prezzo: {cibo['prezzo']}")
            print(f"Quantità: {cibo['quantità']}")
            print(f"Scadenza: {cibo['scadenza']}")
            print(f"Condizione: {cibo['condizione']}")

# Funzione ricerca elementi sia con uno che con più criteri di ricerca
def ricerca_elementi():
    # Lista di attributi validi
    attributi_validi = ['nome', 'categoria', 'prezzo', 'quantità', 'scadenza', 'condizione']

    # Input per capire che tipo di filtraggio vuole fare l'utente
    scelta = input('Vuoi fare una ricerca con uno o più attributi? (uno/multipli): ').lower()

    if scelta == 'uno':

        # Input per la ricerca semplice
        attributo = input("Inserisci l'attributo per la ricerca (nome/categoria/prezzo/quantità/scadenza/condizione): ").lower()
        valore = input(f"Inserisci il valore per l'attributo {attributo}: ")

        # Controllo validità attributo
        if attributo not in attributi_validi:
            print('Errore! Attributo non presente')
            return

        # Conversione del valore se necessario
        if attributo in ['prezzo', 'quantità']:
            try:
                if attributo == 'prezzo':
                    valore = float(valore)
                else:
                    valore = int(valore)
            except ValueError:
                print(f'Errore! {attributo} deve essere un numero valido (int/float).')
                return

        # Ricerca degli elementi corrispondenti
        risultati = [cibo for cibo in elementi if cibo.get(attributo) == valore]

    elif scelta == 'multipli':
        #lista criteri di ricerca
        criteri = []
        risultati = []

        while True:
            attributo = input('Inserisci un attributo per la ricerca (nome/categoria/prezzo/quantità/scadenza/condizione): ').lower()
            valore = input(f"Inserisci un valore per l'attributo {attributo}: ")

            if attributo not in attributi_validi:
                print('Errore, attributo non presente')
                continue

            if attributo in ['prezzo', 'quantità']:

                try:
                    if attributo == 'prezzo':
                        valore = float(valore)
                    else:
                        valore = int(valore)
                except ValueError:
                    print(f'Errore! {attributo} deve essere un numero valido (int/float).')
                    continue

            # Aggiungo la tupla (attributo, valore) alla lista dei criteri
            criteri.append((attributo, valore))

            # Chiedo all'utente se vuole aggiungere un altro criterio
            altro = input('Vuoi aggiungere un altro valore di ricerca? (si/no): ').lower()
            if altro != 'si':
                break

        # Filtraggio avanzato con più criteri
        for cibo in elementi:
            corrisponde = True # Variabile booleana che assume inizialmente che l'elemento sia valido
            for attributo, valore in criteri:
                if cibo[attributo] != valore:
                    corrisponde = False
                    break

            if corrisponde:
                risultati.append(cibo)

    else:
        print('Scelta non valida. Riprova')
        return

    # Stampa dei risultati
    if risultati:
        print('\nElementi trovati:')
        for cibo in risultati:
            print(f"\nNome: {cibo['nome']}")
            print(f"Categoria: {cibo['categoria']}")
            print(f"Prezzo: {cibo['prezzo']}")
            print(f"Quantità: {cibo['quantità']}")
            print(f"Scadenza: {cibo['scadenza']}")
            print(f"Condizione: {cibo['condizione']}")
    else:
        print('Nessun elemento trovato.')


def statistiche_elementi():
    if not elementi: # controllo se la lista sia vuota o meno
        print('Nessun elemento registrato')
        return

    # conteggio totale elementi
    totale_elementi = len(elementi)
    print(f'Il totale degli elementi registrati è {totale_elementi}')

    # conteggio per categoria
    conta_categorie = {} # dizionario vuoto dove andrà la coppia categoria : numero

    for elemento in elementi:
        categoria = elemento.get('categoria', 'sconosciuta') # comando che cerca la chiave categoria nel dizionario e se non c'è restituisce 'sconosciuta'

        # aggiorniamo il dizionario. La chiave categoria del dizionario si aggiorna prendendo la variabile categoria e aggiungendo 1
        # Altrimenti se non esiste metterà 0
        conta_categorie[categoria] = conta_categorie.get(categoria, 0) + 1

    print('\nDistribuzione per categoria: ')
    for categoria, conteggio in conta_categorie.items(): # .items() per iterare sulla coppia del dizionario
        print(f'{categoria}, {conteggio}')

    # categoria più frequente
    categoria_piu_frequente = max(conta_categorie, key=conta_categorie.get) # trova il massimo della variabile e key dice alla funzione max usare il valore della chiave come riferimento del confronto
    print(f'\nCategoria più frequente: {categoria_piu_frequente}, {conta_categorie[categoria_piu_frequente]} elementi')

    # cibo più presente
    def max_quantita(elemento):
        return elemento['quantità']

    cibo_piu_presente = max(elementi, key=max_quantita) # key come prima è il criterio di confronto per max
    print(f'\nIl cibo più presente è: {cibo_piu_presente["nome"]}, {cibo_piu_presente["quantità"]}')

    # statistiche usi prezzi
    prezzi = [e['prezzo'] for e in elementi] # list comprehension che prende solo il valore della chiave prezzo in elementi

    prezzo_massimo = max(prezzi)
    prezzo_minimo = min(prezzi)
    prezzo_medio = sum(prezzi) / len(prezzi)

    # Stampo le statistiche sui prezzi
    print('\nStatistiche sui prezzi:')
    print(f'Il prezzo massimo è {prezzo_massimo:.2f}') #.2f serve per far apparire i prezzi con due cifre decimali
    print(f'Il prezzo minimo è {prezzo_minimo:.2f}')
    print(f'Il prezzo medio è {prezzo_medio:.2f}')

def elimina_elemento():
    nome = input('Inserisci il nome dell\'elemento da eliminare: ')
    elementi[:] = [e for e in elementi if e['nome'].lower() != nome.lower()]  # i ':' indicano che la modifica è fatta sulla lista stessa e non viene creata un'altra
    print(f"Elemento '{nome}' eliminato con successo.")

def menu():
    while True:
        print("\n--- Sistema di Gestione Elementi ---")
        print("1. Aggiungi elemento")
        print("2. Visualizza elementi")
        print("3. Ricerca elemento")
        print("4. Statistiche elementi")
        print("5. Elimina elemento")
        print("6. Esci")

        scelta = input("Scegli un'opzione: ")

        if scelta == "1":
            inserimento_elementi()
        elif scelta == "2":
            visualizza_elementi()
        elif scelta == "3":
            ricerca_elementi()
        elif scelta == "4":
            statistiche_elementi()
        elif scelta == "5":
            elimina_elemento()
        elif scelta == "6":
            print("Uscita dal sistema.")
            break
        else:
            print("Scelta non valida. Riprova.")

if __name__ == '__main__': # serve per assicurarsi che il programma si avvii solo quando è eseguito direttamente e non importato in un altro script
    menu()
```
